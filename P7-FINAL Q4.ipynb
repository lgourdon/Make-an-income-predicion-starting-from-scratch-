#### Régression linéaire et summary
* Nous avons une variable qualitative (country) et une variable quantitative (income_c)
* Nous utiliserons donc la méthode ANOVA pour notre démonstration

```python
# regression OLS
anova = smf.ols('income_c~country', data=df1).fit()
print(anova.summary())
```

```python
# Analyse de la variance
aov_table = sm.stats.anova_lm(anova, typ=1)
aov_table
```

>* R-squared = 38.8% variance expliquée
>* Résidus = 61.2% variance non expliquée

### Test de Fisher
Loi de [Fisher](https://www.youtube.com/watch?v=y8nRhsixBPs) : on détermine la F-value à partir de la somme des carrés intra-classes et inter-classes et de leurs degrés de liberté.

* H0 = le revenu moyen par centile pays est égal (le pays n'a pas d'influence sur le revenu )
* H1 = le revenu moyen par centile pays est différent (au moins 2 pays ont des revenus moyens par centiles différents)<br><br>

* Seuil de signification **α = 0.05**

<br><br>[**Calcul de la valeur F critique**](http://www.statdistributions.com/f?p=0.05&df1=2&df2=4) pour **α** = 0.05
![](img/F-value2.png)

>* Pour **α = 0.05** la F-value est de **6.944**. Notre F-value étant de **73.30**, l'hypothèse H0 est rejetée.<br>
>* Il y a bien un lien entre income_c et country : au moins 2 pays ont des revenus moyens par centiles différents

### Diagnostic du modèle paramétrique

#### Normalité des résidus
<span style="color:#454545">**Test de Shapiro-Wilk**</span><br>
*L'hypothèse nulle démontre que la population est normalement distribuée*
* Si la p-value est **inférieure à α** l'hypothèse nulle est rejetée
* Si la p-value est **supérieure à α** on ne peut pas la rejeter<br><br>

* Seuil de signification **α = 0.05**

```python
# Test de Shapiro-Wilk sous pingouin
pg.normality(anova.resid)
```

>* La p-value est nulle à deux décimales, donc en-dessous du seuil de signification. 
>* L'hypothèse nulle est rejetée. Pas de distribution normale (gaussienne) a priori

<span style="color:#454545">**Vérification graphique**</span><br>
*Normal Q-Q plot*

```python
# Q-Q plot résiduals studentisés
model_norm_residuals = anova.get_influence().resid_studentized_internal
sns.set_style("ticks")
fig, ax = plt.subplots(figsize=[18,5])

plt.subplot(1, 2, 1)
plt.grid(linestyle=':', color='black',linewidth=0.5)
stats.probplot(model_norm_residuals, plot=plt, fit=True)
plt.title("Normal Q-Q", fontsize=20, color='navy')
plt.xlabel('Theoretical Quantiles', fontsize=14)
plt.ylabel('Standardized Residuals', fontsize=14)

plt.savefig('img/4.2 QQplot.png', dpi=120, bbox_inches='tight')
plt.xlim((-3.5,3.5))
plt.show()
```

>* Le Q-Q plot vient confirmer graphiquement l'hypothèse de non-normalité

#### Homoscédasticité

<span style="color:#454545">**Test de Levene**</span><br>
Le test de Levene évalue si les variances de population sont homogènes<br>
*L'hypothèse nulle démontre que la population est normalement distribuée*

* Si la p-value est inférieure au seuil de signification **α**, l'hypothèse nulle est rejetée (il existe une différence entre les variances de population)
* Si la p-value est supérieure au seuil de signification **α**, on ne doit pas rejeter l'hypothèse nulle : les variances de population sont égales (homoscédasticité)<br><br>

* Seuil de signification **α = 0.05**

```python
# Test de Levene sous pingouin
lv=df1[['country','income_c']].copy()
pg.homoscedasticity(lv, dv='income_c', group='country')
```

>* La p-value est nulle à deux décimales, donc sous du seuil de signification, **l'hypothèse nulle est rejetée**
>* Il existe une différence entre les variances de population d'au moins deux pays

<span style="color:#454545">**Représentation graphique des residuals**</span><br>

```python
# fitted values (constante pour l'intercept)
model_fitted_y = anova.fittedvalues
# normalized residuals
model_norm_residuals = anova.get_influence().resid_studentized_internal
# absolute squared normalized residuals
model_norm_residuals_abs_sqrt = np.sqrt(np.abs(model_norm_residuals))

# représentation graphique des squared standardized residuals
sns.set_style("ticks")
fig, ax = plt.subplots(figsize=[16,10])

plt.subplot(221)
plt.scatter(model_fitted_y, model_norm_residuals_abs_sqrt, alpha=0.5)
sns.regplot(model_fitted_y, model_norm_residuals_abs_sqrt, 
            scatter=False, ci=False, lowess=True,
            line_kws={'color': 'red', 'lw': 1, 'alpha': 0.8})

plt.grid(linestyle=':', color='black',linewidth=0.5)
plt.title('Scale-Location', size=20, color="navy")
plt.xlabel('Fitted values', fontsize=14)
plt.ylabel('$\sqrt{|Standardized Residuals|}$', fontsize=14);

plt.savefig('img/4.2 homoscedasticite.png', dpi=120, bbox_inches='tight')
plt.show()
```

>* Plus valeurs prédites augmentent, plus l'écart entre valeurs prédites et observées augmente
>* Plus les valeurs augmentent, plus la ligne augmente
>* Nous rejetons l'hypothèse de l'homogénéité des variances

## ANOVA-One Way non paramétrique [country~income_c]
Dans cette situation de rejet des hypothèses de normalité et d’homogénéité des résidus, il est possible de comparer les médianes à l’aide d’une ANOVA non paramétrique, grâce au test de Kruskal-Wallis<br><br>
Les différences entre les médianes sont statistiquement significatives, en comparant la p-value au seuil de signification pour évaluer l'hypothèse nulle. 

* H0 = les médianes de population sont toutes égales
* H1= aux moins deux médianes de population sont différentes<br><br>

* Seuil de signification **α = 0.05**

### Diagnostics du modèle

#### Normalité des résidus

<span style="color:#454545">**Test de Shapiro-Wilk**</span><br>
*L'hypothèse nulle démontre que la population est normalement distribuée*
* Si la p-value est **inférieure à α** l'hypothèse nulle est rejetée
* Si la p-value est **supérieure à α** on ne peut pas la rejeter<br><br>

* Seuil de signification **α = 0.05**

```python
# Test de Shapiro-Wilk sous pingouin
pg.normality(lr1.resid)
```

>* La p-value est nulle à deux décimales, donc en-dessous du seuil de signification. 
>* L'hypothèse nulle est rejetée. Nous ne pouvons pas conclure à priori à une distribution normale (gaussienne)

*Normal Q-Q plot*

```python
# Q-Q plot résiduals studentisés
model_norm_residuals = lr1.get_influence().resid_studentized_internal

sns.set_style("ticks")
fig, ax = plt.subplots(figsize=[18,5])
plt.subplot(1, 2, 1)
plt.grid(linestyle=':', color='black',linewidth=0.5)

stats.probplot(model_norm_residuals, plot=plt, fit=True)
plt.title("Normal Q-Q", fontsize=20, color='navy')
plt.xlabel('Theoretical Quantiles',fontsize=14)
plt.ylabel('Standardized Residuals',fontsize=14)
plt.xlim((-3.5,3.5))

plt.subplot(1, 2, 2)
sns.distplot(lr1.resid)
plt.grid(linestyle=':', color='black',linewidth=0.5)
plt.title("Histogramme Residuals", size=20, color="navy")

plt.savefig('img/4.4.QQplot.png', dpi=120, bbox_inches='tight')
plt.show()
```

>* L'histogramme des résidus à droite montre que les individus ne sont pas symétriquement distribués autour de l'axe 0 du fait d'un fort skewness à droite

#### Homoscédasticité

<span style="color:#454545">**Test de Levene**</span><br>
Le test de Levene évalue si les variances de population sont homogènes<br>
*L'hypothèse nulle démontre que la population est normalement distribuée*

* Si la p-value est inférieure au seuil de signification **α**, l'hypothèse nulle est rejetée (il existe une différence entre les variances de population)
* Si la p-value est supérieure au seuil de signification **α**, on ne doit pas rejeter l'hypothèse nulle : les variances de population sont égales (homoscédasticité)<br><br>

* Seuil de signification **α = 0.05**

```python
# Test de Levene sous pingouin
lv=df1[['country','income_c','gdp_pc','gini']].copy()
pg.homoscedasticity(lv, dv='income_c', group='country')
```

>* La p-value est nulle, donc sous du seuil de signification, **l'hypothèse nulle est rejetée**
>* Il existe a priori une différence entre les variances de population

<span style="color:#454545">**Représentation graphique**</span>

```python
# fitted values (constante pour l'intercept)
model_fitted_y = lr1.fittedvalues
# normalized residuals
model_norm_residuals = lr1.get_influence().resid_studentized_internal
# absolute squared normalized residuals
model_norm_residuals_abs_sqrt = np.sqrt(np.abs(model_norm_residuals))

# représentation graphique des suqared standardized residuals
sns.set_style("ticks")
fig, ax = plt.subplots(figsize=[16,10])

plt.subplot(221)
plt.scatter(model_fitted_y, model_norm_residuals_abs_sqrt, alpha=0.5)
sns.regplot(model_fitted_y, model_norm_residuals_abs_sqrt, 
            scatter=False, 
            ci=False, 
            lowess=True,
            line_kws={'color': 'red', 'lw': 2, 'alpha': 0.8})

plt.grid(linestyle=':', color='black',linewidth=0.5)
plt.title('Scale-Location', size=20, color="navy")
plt.xlabel('Fitted values', fontsize=14)
plt.ylabel('$\sqrt{|Standardized Residuals|}$', fontsize=14)

plt.savefig('img/4.4.homoscedasticite.png', dpi=120, bbox_inches='tight')
plt.show()
```

>* Plus valeurs prédites augmentent, plus l'écart entre valeurs prédites et observées augmente
>* Plus les valeurs augmentent, plus la ligne augmente
>* Nous rejetons l'hypothèse de l'homogénéité des variances

### Diagnostics du modèle

#### Normalité des résidus
<span style="color:#454545">**Test de Shapiro-Wilk**</span><br>
*L'hypothèse nulle démontre que la population est normalement distribuée*
* Si la p-value est **inférieure à α** l'hypothèse nulle est rejetée
* Si la p-value est **supérieure à α** on ne peut pas la rejeter<br><br>

* Seuil de signification **α = 0.05**

```python
# Test de Shapiro-Wilk sous pingouin
pg.normality(lr4.resid)
```

>* La p-value est nulle à deux décimales, donc en-dessous du seuil de signification. 
>* L'hypothèse nulle est rejetée. Nous ne pouvons pas conclure à priori à une distribution normale (gaussienne)

<span style="color:#454545">**Vérification graphique**</span><br>
*Normal Q-Q plot*

```python
# Q-Q plot résiduals studentisés
model_norm_residuals = lr4.get_influence().resid_studentized_internal
fig, ax = plt.subplots(figsize=[18,5])

plt.subplot(1, 2, 1)
plt.plot(lr4.predict(), lr4.resid,'.', alpha=0.4)
plt.grid(linestyle=':', color='black',linewidth=0.5)
stats.probplot(model_norm_residuals, plot=plt, fit=True)
plt.title("Normal Q-Q", fontsize=20, color='navy')
plt.xlabel('Theoretical Quantiles', fontsize=14)
plt.ylabel('Standardized Residuals', fontsize=14)
plt.xlim((-4.5,4.5))

plt.subplot(1, 2, 2)
sns.distplot(lr4.resid)
plt.grid(linestyle=':', color='black',linewidth=0.5)
plt.title("Histogramme Residuals", size=20, color="navy")

plt.savefig('img/7.4.QQplot.png', dpi=120, bbox_inches='tight')
plt.show()
```

>* Le Q-Q plot montre cependant une distribution normale pour les quantiles théoriques situés entre -2.5 et +2.5
>* L'histogramme des résiduals est plutôt bien réparti autour de l'axe 0
>* Nous pouvons donc considérer que les résidus suivent une distribution gaussienne

#### Homoscédasticité

<span style="color:#454545">**Test de Levene**</span><br>
Le test de Levene évalue si les variances de population sont homogènes<br>
*L'hypothèse nulle démontre que la population est normalement distribuée*

* Si la p-value est inférieure au seuil de signification **α**, l'hypothèse nulle est rejetée (il existe une différence entre les variances de population)
* Si la p-value est supérieure au seuil de signification **α**, on ne doit pas rejeter l'hypothèse nulle : les variances de population sont égales (homoscédasticité)<br><br>

* Seuil de signification **α = 0.05**

```python
# Test de Levene sous pingouin
lv=df4[['country','logincome_c','classe_p','loggdp_pc','loggini']].copy()
pg.homoscedasticity(lv, dv='logincome_c', group='country')
```
